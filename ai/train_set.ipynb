```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
train.head()
```

```python
test.head()
```

```python
train = train[['감정_대분류','사람문장1','사람문장2','사람문장3','사람문장4']]
test = test[['감정_대분류','사람문장1','사람문장2','사람문장3','사람문장4']]
```

```python
train.head()
```

```python
train['감정_대분류'].unique()
```

```python
emotion = {'기쁨':0,'불안':1,'당황':2,'슬픔':3,'분노':4,'상처':5}
```

```python
for i in range(len(train['감정_대분류'])):
  train['감정_대분류'][i] = emotion[train['감정_대분류'][i]]
```

```python
for i in range(len(test['감정_대분류'])):
  test['감정_대분류'][i] = emotion[test['감정_대분류'][i]]
```

```python
train
```

```python
train = train.fillna('')
test = test.fillna('')
```

```python
train['문장'] = ''
test['문장'] = ''
```

```python
train
```

```python
for i in range(len(train)):
  train['문장'][i] = train['사람문장1'][i]+train['사람문장2'][i]+train['사람문장3'][i]+train['사람문장4'][i]
for i in range(len(test)):
  test['문장'][i] = test['사람문장1'][i]+test['사람문장2'][i]+test['사람문장3'][i]+test['사람문장4'][i]
```

```python
train_set = train[['감정_대분류','문장']]
test_set = test[['감정_대분류','문장']]
```

```python
train_set
```

```python
train_set.to_csv('train_set.txt', sep = '\t', index = False)
test_set.to_csv('test_set.txt', sep = '\t', index = False)
```

```python
train_set = train_set[train_set['감정_대분류'] != 0][train_set['감정_대분류'] != 1][train_set['감정_대분류'] != 2][train_set['감정_대분류'] != 4]
```

```python
test_set = test_set[test_set['감정_대분류'] != 0][test_set['감정_대분류'] != 1][test_set['감정_대분류'] != 2][test_set['감정_대분류'] != 4]
```
